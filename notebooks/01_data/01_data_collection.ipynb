# Setup and Imports

```python
import sys
import os
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

# Set project root
PROJECT_ROOT = Path.cwd().parent.parent
sys.path.append(str(PROJECT_ROOT))

print("✅ Setup complete")
print(f"📁 Project root: {PROJECT_ROOT}")

# Cell 2: Data Collection
class DataCollector:
    def fetch_sp500_data(self, start_date='2010-01-01', end_date=None):
        if end_date is None:
            end_date = datetime.now().strftime('%Y-%m-%d')
        
        print(f"📊 Fetching S&P 500 data from {start_date} to {end_date}")
        ticker = yf.Ticker("^GSPC")
        df = ticker.history(start=start_date, end=end_date)
        df.columns = [col.lower() for col in df.columns]
        df['returns'] = df['close'].pct_change()
        df['target'] = df['close'].shift(-5) / df['close'] - 1
        df = df.dropna()
        
        print(f"✅ Fetched {len(df)} rows")
        return df

# Collect data
collector = DataCollector()
df_raw = collector.fetch_sp500_data()
print(f"\n📊 Data shape: {df_raw.shape}")
print(df_raw.head())

# Cell 3: Save Raw Data
# Save to raw directory
raw_path = PROJECT_ROOT / 'data/raw/yfinance/sp500_raw.csv'
raw_path.parent.mkdir(parents=True, exist_ok=True)
df_raw.to_csv(raw_path)
print(f"✅ Raw data saved to {raw_path}")

# Cell 4: Data Validation
print("\n🔍 Data Validation:")
print(f"   Missing values: {df_raw.isnull().sum().sum()}")
print(f"   Duplicates: {df_raw.index.duplicated().sum()}")
print(f"   Date range: {df_raw.index[0]} to {df_raw.index[-1]}")
print(f"   Columns: {list(df_raw.columns)}")
```

# Setup - Notebook Compatible

```python
import os
import sys
import subprocess
from pathlib import Path

# Set project root (notebook-compatible way)
PROJECT_ROOT = Path(os.getcwd()).parent
print(f"📁 Project root: {PROJECT_ROOT}")

def run_python_script(script_path, description):
    """Run a Python script"""
    print(f"\n{'='*60}")
    print(f"📊 {description}")
    print(f"{'='*60}")
    
    full_path = PROJECT_ROOT / script_path
    if full_path.exists():
        print(f"Running: {full_path}")
        result = subprocess.run([sys.executable, str(full_path)], 
                              capture_output=True, text=True)
        if result.returncode == 0:
            print("✅ Success")
            if result.stdout:
                print(result.stdout[-500:])  # Last 500 chars
        else:
            print(f"❌ Failed with error:")
            print(result.stderr)
    else:
        print(f"⚠️ Script not found: {full_path}")

def run_notebook(notebook_path, description):
    """Run a Jupyter notebook"""
    print(f"\n{'='*60}")
    print(f"📓 {description}")
    print(f"{'='*60}")
    
    full_path = PROJECT_ROOT / notebook_path
    if full_path.exists():
        print(f"📓 Notebook: {full_path}")
        print("💡 To run this notebook, open it in Jupyter:")
        print(f"   jupyter notebook {full_path}")
        print("   Then run all cells manually")
    else:
        print(f"⚠️ Notebook not found: {full_path}")
```

# Run Available Scripts

```python
print("\n🎨 RUNNING ALL VISUALIZATIONS AND REPORTS")
print("="*60)

# 1. Generate Report
run_python_script("scripts/generate_report.py", "Generating Performance Report")

# 2. SHAP Analysis (Notebook - manual)
run_notebook("src/visualization/plots/shap_summary.ipynb", "SHAP Analysis")
```
